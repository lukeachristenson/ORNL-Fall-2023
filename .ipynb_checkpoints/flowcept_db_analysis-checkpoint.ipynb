```python
def task_query(_filter: dict, telemetry_keys: List[str] = None, 
                        only_best_task: bool = False):
    # Define the projection to select specific fields
    projection = {
        "_id": 1,
        "workflow_id": 1,
        "task_id": 1,
        "started_at": 1,
        "utc_timestamp": 1,
        "submitted_at": 1,
        "ended_at": 1,
        "node_name": 1,
        "used": 1,
        "generated": 1,
        "stderr": 1,
        "stdout": 1,
        # "error": 1,
        # "finished": 1,
        # "flowcept_version": 1,
        # "telemetry_at_start": 1,
        # "telemetry_at_end": 1
    }
    
    if telemetry_keys is not None:
        _filter.update({"telemetry_at_start": {"$exists": True}, "telemetry_at_end": {"$exists": True}})
        projection.update({"telemetry_at_start": 1, "telemetry_at_end": 1})

    results = collection.find(_filter, projection)

    # Initialize variables for best task and documents
    best_task = None
    highest_accuracy = -1
    docs = []
    new_hp_data = []

    for doc in results:
        # Process telemetry
        if telemetry_keys:
            telemetry_at_start = doc.get("telemetry_at_start", {})
            telemetry_at_end = doc.get("telemetry_at_end", {})
            telemetry_diff = calculate_diff(telemetry_at_start, telemetry_at_end)
            doc['telemetry_diff'] = telemetry_diff

        # Process best task logic
        if only_best_task:
            current_accuracy = doc.get("generated", {}).get("final_evaluation_accuracy", -1)
            if current_accuracy > highest_accuracy:
                highest_accuracy = current_accuracy
                best_task = doc
        else:
            docs.append(doc)

        # Process hyperparameters
        # if hp_terms:
        #     new_hp_row = process_hyperparameters(doc, hp_terms)
        #     new_hp_data.append(new_hp_row)

    # Return the best task if only_best_task is True
    if only_best_task:
        return best_task

    # Convert documents to DataFrame
    df = pd.json_normalize(docs)
    df['utc_timestamp'] = pd.to_datetime(df['utc_timestamp'], unit='s') + timedelta(hours=-8)
    try:
        df['started_at'] = pd.to_datetime(df['started_at'], unit='s') + timedelta(hours=-8)
        df['submitted_at'] = pd.to_datetime(df['submitted_at'], unit='s') + timedelta(hours=-8)
        df['ended_at'] = pd.to_datetime(df['ended_at'], unit='s') + timedelta(hours=-8)
    except:
        pass

    df['doc_generated_time'] = df['_id'].apply(lambda _id: ObjectId(_id).generation_time + timedelta(hours=-4))

    try:
        df['elapsed_time'] = df['ended_at'] - df['started_at']
        df['elapsed_time'] = df['elapsed_time'].apply(lambda x: x.total_seconds() if isinstance(x, timedelta) else -1) 
    except:
        pass

    # If hyperparameter terms were processed, add them to the DataFrame
    if hp_terms:
        df_hpsettings = pd.DataFrame(new_hp_data)
        df = pd.concat([df, df_hpsettings], axis=1)

    return df

def calculate_diff(start, end):
    diff = {}
    #Iterating over keys in start, keys in start and end are the same
    for key in start:
        if isinstance(start[key], Mapping) and isinstance(end.get(key), Mapping):
            # If the value is another dictionary, recursively calculate the difference
            diff[key] = calculate_diff(start[key], end[key])
        elif isinstance(start[key], (int, float)) and isinstance(end.get(key), (int, float)):
            # If the value is a number, calculate the difference directly
            diff[key] = end[key] - start[key]
        else:
            # I'm not sure what should happen in the else case because I'm unsure of the structure
            diff[key] = None
    return diff

#this makes it so that the 
# def process_hyperparameters(doc, hp_terms):
#     # Function logic to process hyperparameters
#     new_hp_row = {}
#     for term, new_key in hp_terms.items():
#         pattern = rf'used[.]{term}'
#         term_data = doc.filter(regex=pattern)
#         if term_data.any():
#             new_hp_row[new_key] = list(set(term_data))[0]
#     return new_hp_row

# Usage example
# client = MongoClient('mongodb://localhost:27017/')
# db = client['your_db_name']
# collection = db['your_collection_name']
# after_date = ...
# filter_criteria = {
#     "ended_at": {"$gte": after_date},
#     "finished": True
# }
# hp_terms = {...} # Define your hyperparameter terms here
# df = combined_task_query(collection, filter_criteria, telemetry_keys=["..."], 
#                          only_best_task=False, hp_terms=hp_terms)
```

```python
import pandas as pd

def extract_information(task_doc, workflow_doc):
    info_dict = {}

    # Extracting information from task document
    info_dict['conda_environment'] = task_doc.get('used', {}).get('sys_name', '')
    info_dict['flowcept_version'] = workflow_doc.get('flowcept_version', '')
    info_dict['hpc_environment'] = task_doc.get('extra_metadata', {}).get('machine', '')
    info_dict['job_directory'] = task_doc.get('used', {}).get('node_name', '')
    info_dict['node'] = task_doc.get('used', {}).get('node_name', '')
    info_dict['gpu_device'] = task_doc.get('generated', {}).get('cuda_visible_device', '')
    info_dict['trainer_username'] = task_doc.get('user', '')
    info_dict['submitted_at'] = pd.to_datetime(task_doc.get('extra_metadata', {}).get('timestamp'), errors='coerce').strftime('%Y-%m-%d %H:%M')
    info_dict['started_at'] = pd.to_datetime(task_doc.get('extra_metadata', {}).get('start_time'), errors='coerce').strftime('%Y-%m-%d %H:%M')
    info_dict['ended_at'] = pd.to_datetime(task_doc.get('generated', {}).get('end_time'), errors='coerce').strftime('%Y-%m-%d %H:%M')
    info_dict['elapsed_time'] = task_doc.get('generated', {}).get('end_time', 0) - task_doc.get('extra_metadata', {}).get('start_time', 0)
    info_dict['final_evaluation_accuracy'] = task_doc.get('generated', {}).get('final_evaluation_accuracy', 0)
    info_dict['final_evaluation_loss'] = task_doc.get('generated', {}).get('final_evaluation_loss', 0)
    info_dict['ref_id'] = task_doc.get('task_id', '')
    info_dict['model_path'] = task_doc.get('generated', {}).get('model_path', '')
    # info_dict['telemetry_diff'] = str(task_doc.get('telemetry_diff', {}))  # Convert to string if needed

    # Extracting information from workflow document
    info_dict['campaign_description'] = workflow_doc.get('comment', '')
    # info_dict['training_description'] = workflow_doc.get('custom_metadata', {}).get('graphene_dataset_info', '')
    # cannot extract name of the dataset because the metadata is custom for each workflow.
    info_dict['training_dataset_path'] = workflow_doc.get('hpsettings', {}).get('training_dataset', '')
    info_dict['test_size'] = workflow_doc.get('hpsettings', {}).get('test_sizes', [None])[0]

    # Additional fields can be added as necessary

    return info_dict
```

```python
def generate_html(template_path, output_directory, info_dict):
    with open(template_path) as f:
        file_content = f.read()

    # Iterate over each key-value pair in the dictionary and replace in the HTML template
    for key, value in info_dict.items():
        # Handle None values to avoid replacing with the string "None"
        if value is None:
            value = ""
        placeholder = "${" + key + "}"
        file_content = file_content.replace(placeholder, str(value))

    # Define the output path using the task_id from info_dict
    task_id = info_dict.get("task_id", "default_task_id")
    output_path = os.path.join(output_directory, f"{task_id}.html")
    
    with open(output_path, "w+") as f:
        f.write(file_content)

# Example usage
# generate_html("/path/to/template.html", "/path/to/output/directory", best_task_info)
```
